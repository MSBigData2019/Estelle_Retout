```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
import json
import re
```

```python
req = requests.get("https://www.open-medicaments.fr/api/v1/medicaments?query=paracetamol")
df = pd.read_json(req.content)
table = df.denomination.str.extract(r'(\D*)(\d+) (\w+), (\w+)')
table["codeCIS"] = df.codeCIS
```

```python
table.columns = ["Name", "Dosage", "Unit", "type", "codeCIS"]
```

```python
table['Dosage'] = table['Dosage'].where(table["Unit"]=="mg", table["Dosage"]+"000")
table['Unit'] = "mg"
```
